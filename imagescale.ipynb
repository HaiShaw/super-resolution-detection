```python
%pylab inline
from PIL import Image
from os import listdir
from os.path import isfile,join
import matplotlib.pyplot as plt

image_dir = "./dataset/VOC12-HR"
image_target_dir = "./dataset/VOC12-LR-X8"

file_name = "2007_001872.jpg"
full_name = image_dir + "/" + file_name
full_target = image_target_dir + "/" + file_name

def scale_down_one_file(input_file,output_file,scale=8):
            img = Image.open(input_file)
            plt.imshow(img)
            plt.show()  
            img = img.resize((img.width//scale,img.height//scale),Image.BICUBIC)
            plt.imshow(img)
            plt.show()            
            img.save(output_file,"JPEG",quality=90)
            img.close()

scale_down_one_file(full_name,full_target)
```

```python
%pylab inline
from PIL import Image
from os import listdir
from os.path import isfile,join
import matplotlib.pyplot as plt

image_dir = "./dataset/VOC12-HR"
image_target_dir = "./dataset/VOC12-LR-X8"

file_name = "2007_000027.jpg"
full_name = image_dir + "/" + file_name
full_target = image_target_dir + "/" + file_name

def scale_down(image_dir,image_target_dir,scale=8):
    for i, f in enumerate(listdir(image_dir)):
        if isfile(join(image_dir,f)):
            img = Image.open(join(image_dir,f))
            img = img.resize((img.width//scale,img.height//scale),Image.BICUBIC)
            img.save(join(image_target_dir,f),"JPEG",quality=90)
            if i == 0:
                print(join(image_dir,f))
                plt.imshow(img)
                plt.show()
            img.close()

scale_down(image_dir,image_target_dir)
```
